```python
import pandas as pd 
import numpy as np 
```

```python
match =pd.read_csv('matches.csv')
match.head()
```

```python
match.columns
```

```python
deliveries = pd.read_csv('deliveries.csv')
deliveries.head(10)
```

```python
deliveries.columns
```

```python
# Create a dictionary for name mapping
team_mapping = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Deccan Chargers': 'Sunrisers Hyderabad',
    'Kings XI Punjab': 'Punjab Kings',
    'Rising Pune Supergiant': 'Rising Pune Supergiants'
}

# Apply to both dataframes
match.replace(team_mapping, inplace=True)
deliveries.replace(team_mapping, inplace=True)

# Keep only the current/consistent teams
consistent_teams = [
    'Sunrisers Hyderabad', 'Mumbai Indians', 'Royal Challengers Bangalore',
    'Kolkata Knight Riders', 'Punjab Kings', 'Chennai Super Kings',
    'Rajasthan Royals', 'Delhi Capitals'
]

match = match[match['team1'].isin(consistent_teams) & match['team2'].isin(consistent_teams)]
deliveries = deliveries[deliveries['batting_team'].isin(consistent_teams) & deliveries['bowling_team'].isin(consistent_teams)]
```

```python
# Calculate cumulative runs per match per innings
deliveries['current_score'] = deliveries.groupby(['match_id', 'inning'])['total_runs'].cumsum()

# Calculate cumulative wickets
deliveries['player_dismissed'] = deliveries['player_dismissed'].fillna('0')
deliveries['is_wicket'] = deliveries['player_dismissed'].apply(lambda x: 1 if x != '0' else 0)
deliveries['wickets_fallen'] = deliveries.groupby(['match_id', 'inning'])['is_wicket'].cumsum()

# Calculate current over and ball
deliveries['balls_bowled'] = (deliveries['over'] * 6) + deliveries['ball']
deliveries['balls_left'] = 120 - deliveries['balls_bowled']
```

```python
# Get total runs for each innings
total_score_df = deliveries.groupby(['match_id', 'inning'])['total_runs'].sum().reset_index()
total_score_df.rename(columns={'total_runs': 'final_score'}, inplace=True)

# Merge back to the main deliveries dataframe
final_df = deliveries.merge(total_score_df, on=['match_id', 'inning'])

# Filter out the first 5 overs (optional, but predictions are usually more stable after the powerplay)
final_df = final_df[final_df['over'] >= 5]
```

```python
final_df.head()
```

```python
final_df.columns
```

```python
# Merge the city column from 'match' into your current 'final_df'
# We use 'id' from match and 'match_id' from final_df to link them
final_df = final_df.merge(match[['id', 'city']], left_on='match_id', right_on='id')

# Drop the extra 'id' column created by the merge
final_df.drop(columns=['id'], inplace=True)
```

```python
# Now 'city' is in the index/columns, so this won't crash
final_df = final_df[['batting_team', 'bowling_team', 'city', 'current_score', 'balls_left', 'wickets_fallen', 'final_score']]

# Drop rows where city is missing (NaN)
final_df.dropna(inplace=True)

# Shuffle the data
final_df = final_df.sample(final_df.shape[0])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error

# 1. Define which columns are categorical
trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
], remainder='passthrough')

# 2. Separate Features (X) and Target (y)
X = final_df.drop(columns=['final_score'])
y = final_df['final_score']

# 3. Split the data (80% for training, 20% for testing)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
# Create the pipeline
pipe = Pipeline(steps=[
    ('step1', trf), # Step 1: Encode the teams and cities
    ('step2', LinearRegression()) # Step 2: Apply the algorithm
])

# Train the model
pipe.fit(X_train, y_train)

# Test the model
y_pred = pipe.predict(X_test)

# Check how accurate we are
print(f"R2 Score: {r2_score(y_test, y_pred)}")
print(f"Mean Absolute Error: {mean_absolute_error(y_test, y_pred)}")
```

```python
from xgboost import XGBRegressor

# 1. Update the pipeline to use XGBoost instead of Linear Regression
# We tune some parameters (hyperparameters) to get better results
xg_pipe = Pipeline(steps=[
    ('step1', trf),
    ('step2', XGBRegressor(n_estimators=1000, learning_rate=0.2, max_depth=12, random_state=1))
])

# 2. Train the XGBoost model
xg_pipe.fit(X_train, y_train)

# 3. Predict and evaluate
y_pred_xg = xg_pipe.predict(X_test)

print(f"XGBoost R2 Score: {r2_score(y_test, y_pred_xg)}")
print(f"XGBoost MAE: {mean_absolute_error(y_test, y_pred_xg)}")
```
