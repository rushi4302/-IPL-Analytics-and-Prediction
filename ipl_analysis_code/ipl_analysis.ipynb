```python
# Calculate total runs scored by each batsman
batsman_runs = deliveries.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False)

batsman_runs.head(10)
```

```python
# Calculate number of balls faced by each batsman
balls_faced = deliveries.groupby('batter')['ball'].count()
```

```python
# Calculate strike rate for each batsman
strike_rate = (batsman_runs / balls_faced * 100).round(2)
```

```python
# Combine runs and strike rate into one table
batting_stats = pd.DataFrame({
    'runs': batsman_runs,
    'strike_rate': strike_rate
}).dropna()

batting_stats.sort_values('strike_rate', ascending=False).head(10)
```

```python
# Create a combined batting score using runs and strike rate
batting_stats['impact_score'] = (batting_stats['runs'] * batting_stats['strike_rate']).round(2)

batting_stats.sort_values('impact_score', ascending=False).head(10)
```

```python
# Count number of dismissals (outs) for each batsman
outs = (
    deliveries[
        (deliveries['is_wicket'] == 1) &
        (deliveries['player_dismissed'] != 'None')
    ]
    .groupby('player_dismissed')['is_wicket']
    .count()
)

outs.head(100)
```

```python
# Calculate batting average using total runs and number of outs
batting_stats['outs'] = outs
batting_stats['average'] = (batting_stats['runs'] / batting_stats['outs']).round(2)

batting_stats.sort_values('average', ascending=False).head(10)
```

```python
# Filter batsmen with sufficient runs to avoid small-sample bias
filtered_batting = batting_stats[batting_stats['runs'] >= 1000]

filtered_batting.sort_values('average', ascending=False).head(1000)
```

```python
# Plot economy rate for top bowlers (min 300 overs to avoid small-sample bias)
bowler_stats = (
    deliveries.groupby('bowler')
    .agg(
        runs_conceded=('total_runs', 'sum'),
        balls=('ball', 'count')
    )
)
bowler_stats['overs'] = bowler_stats['balls'] / 6
bowler_stats['economy'] = (bowler_stats['runs_conceded'] / bowler_stats['overs']).round(2)

top_bowlers = bowler_stats[bowler_stats['overs'] >= 300].sort_values('economy').head(10)

import matplotlib.pyplot as plt
top_bowlers['economy'].plot(kind='bar', title='Top Bowlers by Economy Rate (Min 300 Overs)')
plt.ylabel('Economy')
plt.xlabel('Bowler')
plt.tight_layout()
plt.show()
```

```python
# Compare wickets taken vs economy rate for bowlers
bowler_stats = (
    deliveries.groupby('bowler')
    .agg(
        runs_conceded=('total_runs', 'sum'),
        balls=('ball', 'count'),
        wickets=('is_wicket', 'sum')
    )
)

bowler_stats['overs'] = bowler_stats['balls'] / 6
bowler_stats['economy'] = bowler_stats['runs_conceded'] / bowler_stats['overs']

filtered_bowlers = bowler_stats[bowler_stats['overs'] >= 300]

import matplotlib.pyplot as plt
plt.scatter(filtered_bowlers['economy'], filtered_bowlers['wickets'])
plt.xlabel('Economy Rate')
plt.ylabel('Wickets')
plt.title('Wickets vs Economy (Min 300 Overs)')
plt.show()
```

```python
# Create a combined bowler score using wickets, economy, and overs
filtered_bowlers['bowler_score'] = (
    filtered_bowlers['wickets'] / filtered_bowlers['economy']
) * filtered_bowlers['overs']

filtered_bowlers.sort_values('bowler_score', ascending=False).head(10)
```

```python
# Calculate batting runs and balls in death overs (16–20)
death = deliveries[(deliveries['over'] >= 16) & (deliveries['over'] <= 20)]

death_stats = (
    death.groupby('batter')
    .agg(
        runs=('batsman_runs', 'sum'),
        balls=('ball', 'count')
    )
)

death_stats['strike_rate'] = (death_stats['runs'] / death_stats['balls'] * 100).round(2)

death_stats.sort_values('runs', ascending=False).head(10)
```

```python
# Rank batsmen by strike rate in death overs (min 100 balls to avoid small-sample bias)
death_sr = death.groupby('batter').agg(
    runs=('batsman_runs', 'sum'),
    balls=('ball', 'count')
)

death_sr = death_sr[death_sr['balls'] >= 100]
death_sr['strike_rate'] = (death_sr['runs'] / death_sr['balls'] * 100).round(2)

death_sr.sort_values('strike_rate', ascending=False).head(10)
```

```python
# Rank death-overs finishers using combined score (strike rate × runs) with a minimum balls filter
death_finisher = death.groupby('batter').agg(
    runs=('batsman_runs', 'sum'),
    balls=('ball', 'count')
)

death_finisher = death_finisher[death_finisher['balls'] >= 100]
death_finisher['strike_rate'] = (death_finisher['runs'] / death_finisher['balls'] * 100)
death_finisher['finisher_score'] = (death_finisher['runs'] * death_finisher['strike_rate']).round(2)

death_finisher.sort_values('finisher_score', ascending=False).head(10)
```

```python
# Calculate outs per batter vs team
outs_vs_team = (
    deliveries[
        (deliveries['is_wicket'] == 1) &
        (deliveries['player_dismissed'] != 'None')
    ]
    .groupby(['player_dismissed', 'bowling_team'])
    .size()
    .reset_index(name='outs')
    .rename(columns={'player_dismissed': 'batter'})
)

# Merge with runs/balls table
batter_vs_team_full = batter_vs_team.merge(
    outs_vs_team,
    on=['batter', 'bowling_team'],
    how='left'
).fillna({'outs': 0})

# Compute average safely
batter_vs_team_full['average'] = (
    batter_vs_team_full['runs'] /
    batter_vs_team_full['outs'].replace(0, pd.NA)
)

# Filter for reliable samples (e.g., >=300 balls)
batter_vs_team_full = batter_vs_team_full[batter_vs_team_full['balls'] >= 300]

```

```python
import pandas as pd
import numpy as np

# =========================
# LOAD OLD DATA
# =========================
matches = pd.read_csv("matches.csv")
deliveries = pd.read_csv("deliveries.csv")

# =========================
# LOAD 2025 DATA
# =========================
matches_2025 = pd.read_csv("matches2025.csv")
deliveries_2025 = pd.read_csv("deliveries2025.csv")

# =========================
# TEAM CODE → FULL NAME
# =========================
team_map = {
    "RCB": "Royal Challengers Bengaluru",
    "KKR": "Kolkata Knight Riders",
    "CSK": "Chennai Super Kings",
    "MI": "Mumbai Indians",
    "SRH": "Sunrisers Hyderabad",
    "RR": "Rajasthan Royals",
    "DC": "Delhi Capitals",
    "PBKS": "Punjab Kings",
    "LSG": "Lucknow Super Giants",
    "GT": "Gujarat Titans"
}

for col in ["team1", "team2", "toss_winner", "match_winner"]:
    if col in matches_2025.columns:
        matches_2025[col] = matches_2025[col].map(team_map).fillna(matches_2025[col])

# =========================
# FIX DATE (2025)
# =========================
matches_2025["date"] = pd.to_datetime(
    matches_2025["date"],
    format="%B %d,%Y",
    errors="coerce"
)

# =========================
# FORCE SEASON
# =========================
matches_2025["season"] = 2025

# =========================
# RENAME COLUMNS
# =========================
matches_2025 = matches_2025.rename(columns={"match_winner": "winner"})

# =========================
# REMOVE DUPLICATE COLS
# =========================
matches_2025 = matches_2025.loc[:, ~matches_2025.columns.duplicated()]
deliveries_2025 = deliveries_2025.loc[:, ~deliveries_2025.columns.duplicated()]

# =========================
# ADD MISSING MATCH COLS
# =========================
for col in matches.columns:
    if col not in matches_2025.columns:
        matches_2025[col] = np.nan

matches_2025 = matches_2025[matches.columns]

# =========================
# FIX DELIVERIES
# =========================
deliveries_2025["extras_type"] = "None"
deliveries_2025["is_wicket"] = deliveries_2025["player_dismissed"].notna().astype(int)

for col in deliveries.columns:
    if col not in deliveries_2025.columns:
        deliveries_2025[col] = 0

deliveries_2025 = deliveries_2025[deliveries.columns]

# =========================
# APPEND
# =========================
matches_final = pd.concat([matches, matches_2025], ignore_index=True)
deliveries_final = pd.concat([deliveries, deliveries_2025], ignore_index=True)

# =========================
# 🔥 FINAL NUMERIC FIX
# =========================
numeric_cols = [
    "win_by_runs",
    "win_by_wickets",
    "result_margin",
    "target_runs",
    "target_overs",
    "season"
]

for col in numeric_cols:
    if col in matches_final.columns:
        matches_final[col] = pd.to_numeric(
            matches_final[col],
            errors="coerce"
        ).fillna(0)

# =========================
# 🔥 FINAL DATE FIX (TIMESTAMP ERROR KILL)
# =========================
matches_final["date"] = pd.to_datetime(
    matches_final["date"],
    errors="coerce"
)

matches_final = matches_final.dropna(subset=["date"])

# =========================
# SAVE
# =========================
matches_final.to_csv("matches.csv", index=False)
deliveries_final.to_csv("deliveries.csv", index=False)

# =========================
# VERIFY
# =========================
print("MAX SEASON:", matches_final["season"].max())
print("LATEST DATE:", matches_final["date"].max())
print("DATE DTYPE:", matches_final["date"].dtype)
print("DONE ✅")
```
